```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from scipy.optimize import minimize
import numpy as np

# ========== 更大的模擬資料集 ==========
training_data = {
    "000": [1, 0, 1, 0],
    "001": [0, 1, 0, 1],
    "010": [1, 1, 0, 0],
    "011": [0, 0, 1, 1],
    "100": [1, 0, 0, 1],
    "101": [0, 1, 1, 0],
    "110": [1, 1, 1, 0],  # 目標樣本
    "111": [0, 0, 0, 1]
}
target_index = "110"

# ========== 資料電路 U_data ==========
def u_data(data_bits):
    qc = QuantumCircuit(len(data_bits))
    for i, bit in enumerate(data_bits):
        if bit:
            qc.x(i)
    return qc

# ========== 參數化 Oracle (U_L1)，增加複雜度 ==========
def u_l1(params, num_qubits, layers=2):
    qc = QuantumCircuit(num_qubits)
    idx = 0
    for _ in range(layers):
        for i in range(num_qubits):
            qc.ry(params[idx], i)
            idx += 1
        for i in range(num_qubits - 1):
            qc.cz(i, i + 1)
    return qc

# ========== 完整 Grover operator ==========
def grover_operator(params, index_qubits, data_qubits):
    n_idx = len(index_qubits)
    n_data = len(data_qubits)
    qc = QuantumCircuit(n_idx + n_data)

    data_circuit = u_data(training_data[target_index])
    qc.compose(data_circuit, qubits=data_qubits, inplace=True)

    oracle = u_l1(params, n_data)
    qc.compose(oracle, qubits=data_qubits, inplace=True)

    qc.mcx(data_qubits, index_qubits[-1])

    qc.compose(oracle.inverse(), qubits=data_qubits, inplace=True)
    qc.compose(data_circuit.inverse(), qubits=data_qubits, inplace=True)

    diffuser = QuantumCircuit(n_idx)
    diffuser.h(range(n_idx))
    diffuser.x(range(n_idx))
    diffuser.h(n_idx - 1)
    diffuser.mcx(list(range(n_idx - 1)), n_idx - 1)
    diffuser.h(n_idx - 1)
    diffuser.x(range(n_idx))
    diffuser.h(range(n_idx))

    qc.compose(diffuser, qubits=index_qubits, inplace=True)

    return qc

# ========== 定義loss function並用COBYLA最佳化 ==========
index_qubits = list(range(3))
data_qubits = list(range(3, 7))
all_qubits = len(index_qubits) + len(data_qubits)

simulator = Aer.get_backend("qasm_simulator")
init_params = np.random.rand(2 * len(data_qubits)) * np.pi

grover_iterations = int(np.sqrt(len(training_data)))

def loss_func(params):
    qc = QuantumCircuit(all_qubits, len(index_qubits))
    qc.h(index_qubits)

    for _ in range(grover_iterations):
        qc.compose(grover_operator(params, index_qubits, data_qubits), inplace=True)

    qc.measure(index_qubits, range(len(index_qubits)))
    result = execute(qc, backend=simulator, shots=1024).result()
    counts = result.get_counts()
    prob = counts.get(target_index, 0) / 1024

    return 1 - prob

result = minimize(loss_func, init_params, method="COBYLA", options={'maxiter': 30})

print(f"最佳化後loss: {result.fun:.4f}")
print(f"最佳化後參數: {result.x}")

# 最後視覺化結果
qc_final = QuantumCircuit(all_qubits, len(index_qubits))
qc_final.h(index_qubits)
for _ in range(grover_iterations):
    qc_final.compose(grover_operator(result.x, index_qubits, data_qubits), inplace=True)
qc_final.measure(index_qubits, range(len(index_qubits)))

final_result = execute(qc_final, backend=simulator, shots=1024).result()
plot_histogram(final_result.get_counts())
```

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt

# Define binary classification dataset
# Format: feature vector -> label (0 or 1)
original_dataset = {
    "00": 0,
    "01": 0,
    "10": 1,
    "11": 1
}

# Parameters
K = 4  # Size of extended examples
n_qubits_feature = 2  # Number of qubits for features
n_qubits_index = 2    # Number of qubits for index register

# Build extended dataset as described in the paper
def build_extended_dataset(original_data):
    extended_dataset = []
    
    # Group examples by label
    class0 = [k for k, v in original_data.items() if v == 0] #label為0的dataset
    class1 = [k for k, v in original_data.items() if v == 1] #label為1的dataset
    
    # For each example, create an extended example with K-1 opposite label examples
    for key, label in original_data.items():
        D_k = []
        opposite_class = class1 if label == 0 else class0
        
        # Add K-1 examples with opposite label
        for i in range(K-1):
            opposite_example = opposite_class[i % len(opposite_class)]
            D_k.append((opposite_example, 1-label))
        
        # Add the original example as the last pair
        D_k.append((key, label))
        extended_dataset.append(D_k)
    
    return extended_dataset

# Implement U_data: prepares the superposition state
# 首先全部做Hadamard， 再對應feature跟index，只有當feature符合條件時，那個索引才會被影響
def u_data(extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # Prepare superposition for index register
    qc.h(range(n_qubits_index))
    
    # Encode features conditionally based on index
    for i, (feature, _) in enumerate(extended_example):
        # Convert i to binary for index comparison
        bin_i = format(i, f'0{n_qubits_index}b')
        
        # Create control conditions based on index
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)  # Flip for index comparison
        
        # Apply feature encoding conditionally
        for j, bit in enumerate(feature):
            if bit == '1':
                qc.mcx(list(range(n_qubits_index)), n_qubits_index + j)
        
        # Undo index conditions
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)
    
    return qc

# Implement U_L1: trainable variational circuit
def u_l1(params, n_qubits_feature, layers=2):
    qc = QuantumCircuit(n_qubits_feature)
    param_idx = 0
    
    for _ in range(layers):
        # Rotation gates
        for i in range(n_qubits_feature):
            qc.rx(params[param_idx], i)
            param_idx += 1
            qc.ry(params[param_idx], i)
            param_idx += 1
            qc.rz(params[param_idx], i)
            param_idx += 1
        
        # Entanglement
        for i in range(n_qubits_feature-1):
            qc.cx(i, i+1)
        qc.cx(n_qubits_feature-1, 0)  # Make it circular
    
    return qc

# Implement MCZ gate as described in the paper
def mcz_gate(n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # Target index is K-1
    target_index = K-1  
    bin_target = format(target_index, f'0{n_qubits_index}b')
    
    # Apply X gates to the index qubits where target bit is 0
    for i, bit in enumerate(bin_target):
        if bit == '0':
            qc.x(i)
    
    # Apply multi-controlled Z with first feature qubit as control
    controls = list(range(n_qubits_index)) + [n_qubits_index]  # index qubits + first feature qubit
    qc.h(n_qubits_index + n_qubits_feature - 1)  # Apply H before CZ
    qc.mcx(controls, n_qubits_index + n_qubits_feature - 1)
    qc.h(n_qubits_index + n_qubits_feature - 1)  # Apply H after CZ
    
    # Undo X gates
    for i, bit in enumerate(bin_target):
        if bit == '0':
            qc.x(i)
    
    return qc

# Implement U_init (diffusion operator)
def u_init(n_qubits_index):
    qc = QuantumCircuit(n_qubits_index)
    
    # Implement 2|ψ⟩⟨ψ| - I where |ψ⟩ is uniform superposition
    qc.h(range(n_qubits_index))
    qc.x(range(n_qubits_index))
    qc.h(n_qubits_index-1)
    qc.mcx(list(range(n_qubits_index-1)), n_qubits_index-1)
    qc.h(n_qubits_index-1)
    qc.x(range(n_qubits_index))
    qc.h(range(n_qubits_index))
    
    return qc

# Complete U operator as defined in Equation (9)
def u_operator(params, extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # U_data
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # U_L1
    l1_circuit = u_l1(params, n_qubits_feature)
    qc.compose(l1_circuit, qubits=range(n_qubits_index, n_qubits_index + n_qubits_feature), inplace=True)
    
    # MCZ
    qc.compose(mcz_gate(n_qubits_feature, n_qubits_index), inplace=True)
    
    # (U_L1)†
    qc.compose(l1_circuit.inverse(), qubits=range(n_qubits_index, n_qubits_index + n_qubits_feature), inplace=True)
    
    # (U_data)†
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index).inverse(), inplace=True)
    
    # U_init
    qc.compose(u_init(n_qubits_index), qubits=range(n_qubits_index), inplace=True)
    
    return qc

# U_E operator for the last cycle
def u_e_operator(params, extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # U_data
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # U_L1
    l1_circuit = u_l1(params, n_qubits_feature)
    qc.compose(l1_circuit, qubits=range(n_qubits_index, n_qubits_index + n_qubits_feature), inplace=True)
    
    # MCZ
    qc.compose(mcz_gate(n_qubits_feature, n_qubits_index), inplace=True)
    
    # U_init
    qc.compose(u_init(n_qubits_index), qubits=range(n_qubits_index), inplace=True)
    
    return qc

# Calculate loss according to equation (14)
def loss_function(params, extended_example):
    label = extended_example[-1][1]  # Label of last example
    sign_val = 1 if label == 0 else -1  # sign(1/2 - y_k)
    
    # Create circuit for Grover iterations
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index, n_qubits_index + 1)
    
    # Initial state preparation
    qc.h(range(n_qubits_index))
    
    # Grover iterations O(√K)
    grover_iterations = int(np.sqrt(K))
    for i in range(grover_iterations - 1):
        qc.compose(u_operator(params, extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # Last iteration with U_E
    qc.compose(u_e_operator(params, extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # Measure index register and first feature qubit
    qc.measure(range(n_qubits_index), range(n_qubits_index))
    qc.measure(n_qubits_index, n_qubits_index)  # First feature qubit
    
    # Execute circuit
    simulator = Aer.get_backend('qasm_simulator')
    job = execute(qc, simulator, shots=1024)
    counts = job.result().get_counts()
    
    # Check for target index (K-1) with first feature qubit = 1
    target_state = format(K-1, f'0{n_qubits_index}b') + '1'
    prob = counts.get(target_state, 0) / 1024
    
    # Return loss according to equation (14)
    return sign_val * prob

# Training function
def train_gbls(original_dataset, n_epochs=20):
    extended_dataset = build_extended_dataset(original_dataset)
    n_params = 3 * n_qubits_feature * 2  # 3 rotation gates per qubit, 2 layers
    
    # Initialize parameters
    params = np.random.rand(n_params) * 2 * np.pi
    
    training_loss = []
    
    # Training loop
    for epoch in range(n_epochs):
        np.random.shuffle(extended_dataset)
        epoch_loss = 0
        
        for example in extended_dataset:
            # Define the optimization function for this example
            def opt_func(p):
                return loss_function(p, example)
            
            # Update parameters
            result = minimize(opt_func, params, method='COBYLA', options={'maxiter': 5})
            params = result.x
            epoch_loss += result.fun
        
        avg_loss = epoch_loss / len(extended_dataset)
        training_loss.append(avg_loss)
        print(f"Epoch {epoch+1}/{n_epochs}, Loss: {avg_loss:.4f}")
    
    # Plot training loss
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, n_epochs+1), training_loss)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('GBLS Training Loss')
    plt.grid(True)
    plt.show()
    
    return params

# Prediction function as shown in Figure 5
def predict(params, new_input):
    qc = QuantumCircuit(n_qubits_feature, 1)
    
    # Encode input
    for i, bit in enumerate(new_input):
        if bit == '1':
            qc.x(i)
    
    # Apply trained U_L1
    qc.compose(u_l1(params, n_qubits_feature), inplace=True)
    
    # Measure first qubit
    qc.measure(0, 0)
    
    # Execute
    simulator = Aer.get_backend('qasm_simulator')
    job = execute(qc, simulator, shots=1024)
    counts = job.result().get_counts()
    
    # Calculate probability of '1'
    prob_1 = counts.get('1', 0) / 1024
    
    # Return predicted label
    return 1 if prob_1 >= 0.5 else 0

# Train the model
trained_params = train_gbls(original_dataset)

# Test predictions
for input_data, true_label in original_dataset.items():
    pred = predict(trained_params, input_data)
    print(f"Input: {input_data}, True: {true_label}, Predicted: {pred}")
```

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt

# Define binary classification dataset
# Format: feature vector -> label (0 or 1)
original_dataset = {
    "00": 0,
    "01": 0,
    "10": 1,
    "11": 1
}

# Parameters
K = 4  # Size of extended examples
n_qubits_feature = 2  # Number of qubits for features
n_qubits_index = 2    # Number of qubits for index register

# Build extended dataset as described in the paper
def build_extended_dataset(original_data):
    extended_dataset = []
    
    # Group examples by label
    class0 = [k for k, v in original_data.items() if v == 0]  # label為0的dataset
    class1 = [k for k, v in original_data.items() if v == 1]  # label為1的dataset
    
    # For each example, create an extended example with K-1 opposite label examples
    for key, label in original_data.items():
        D_k = []
        opposite_class = class1 if label == 0 else class0
        
        # Add K-1 examples with opposite label
        for i in range(K-1):
            opposite_example = opposite_class[i % len(opposite_class)]
            D_k.append((opposite_example, 1-label))
        
        # Add the original example as the last pair
        D_k.append((key, label))
        extended_dataset.append(D_k)
    
    return extended_dataset

# Implement U_data: prepares the superposition state
# 首先全部做Hadamard， 再對應feature跟index，只有當feature符合條件時，那個索引才會被影響
def u_data(extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # Prepare superposition for index register
    qc.h(range(n_qubits_index))
    
    # Encode features conditionally based on index
    for i, (feature, _) in enumerate(extended_example):
        # Convert i to binary for index comparison
        bin_i = format(i, f'0{n_qubits_index}b')
        
        # Create control conditions based on index
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)  # Flip for index comparison
        
        # Apply feature encoding conditionally
        for j, bit in enumerate(feature):
            if bit == '1':
                qc.mcx(list(range(n_qubits_index)), n_qubits_index + j)
        
        # Undo index conditions
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)
    
    return qc

# Implement U_L1: trainable variational circuit
def u_l1(params, n_qubits_feature, layers=2):
    qc = QuantumCircuit(n_qubits_feature)
    param_idx = 0
    
    for _ in range(layers):
        # Rotation gates
        for i in range(n_qubits_feature):
            qc.rx(params[param_idx], i)
            param_idx += 1
            qc.ry(params[param_idx], i)
            param_idx += 1
            qc.rz(params[param_idx], i)
            param_idx += 1
        
        # Entanglement
        for i in range(n_qubits_feature-1):
            qc.cx(i, i+1)
        qc.cx(n_qubits_feature-1, 0)  # Make it circular
    
    return qc

# Implement MCZ gate as described in the paper
def mcz_gate(n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # Target index is K-1
    target_index = K-1  
    bin_target = format(target_index, f'0{n_qubits_index}b')
    
    # Apply X gates to the index qubits where target bit is 0
    for i, bit in enumerate(bin_target):
        if bit == '0':
            qc.x(i)
    
    # Apply multi-controlled Z with first feature qubit as control
    controls = list(range(n_qubits_index)) + [n_qubits_index]  # index qubits + first feature qubit
    qc.h(n_qubits_index + n_qubits_feature - 1)  # Apply H before CZ
    qc.mcx(controls, n_qubits_index + n_qubits_feature - 1)
    qc.h(n_qubits_index + n_qubits_feature - 1)  # Apply H after CZ
    
    # Undo X gates
    for i, bit in enumerate(bin_target):
        if bit == '0':
            qc.x(i)
    
    return qc

# Implement U_init (diffusion operator)
def u_init(n_qubits_index):
    qc = QuantumCircuit(n_qubits_index)
    
    # Implement 2|ψ⟩⟨ψ| - I where |ψ⟩ is uniform superposition
    qc.h(range(n_qubits_index))
    qc.x(range(n_qubits_index))
    qc.h(n_qubits_index-1)
    qc.mcx(list(range(n_qubits_index-1)), n_qubits_index-1)
    qc.h(n_qubits_index-1)
    qc.x(range(n_qubits_index))
    qc.h(range(n_qubits_index))
    
    return qc

# Complete U operator as defined in Equation (9)
def u_operator(params, extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # U_data
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # U_L1
    l1_circuit = u_l1(params, n_qubits_feature)
    qc.compose(l1_circuit, qubits=range(n_qubits_index, n_qubits_index + n_qubits_feature), inplace=True)
    
    # MCZ
    qc.compose(mcz_gate(n_qubits_feature, n_qubits_index), inplace=True)
    
    # (U_L1)†
    qc.compose(l1_circuit.inverse(), qubits=range(n_qubits_index, n_qubits_index + n_qubits_feature), inplace=True)
    
    # (U_data)†
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index).inverse(), inplace=True)
    
    # U_init
    qc.compose(u_init(n_qubits_index), qubits=range(n_qubits_index), inplace=True)
    
    return qc

# U_E operator for the last cycle
def u_e_operator(params, extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # U_data
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # U_L1
    l1_circuit = u_l1(params, n_qubits_feature)
    qc.compose(l1_circuit, qubits=range(n_qubits_index, n_qubits_index + n_qubits_feature), inplace=True)
    
    # MCZ
    qc.compose(mcz_gate(n_qubits_feature, n_qubits_index), inplace=True)
    
    # U_init
    qc.compose(u_init(n_qubits_index), qubits=range(n_qubits_index), inplace=True)
    
    return qc

# Calculate loss according to binary cross entropy (BCE) loss
def loss_function(params, extended_example):
    # Get label (the label of the last example in extended example)
    label = extended_example[-1][1]  # label in {0,1}
    
    # Create circuit for Grover iterations
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index, n_qubits_index + 1)
    
    # Initial state preparation: apply Hadamard on index register
    qc.h(range(n_qubits_index))
    
    # Grover iterations O(√K)
    grover_iterations = int(np.sqrt(K))
    for i in range(grover_iterations - 1):
        qc.compose(u_operator(params, extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # Last iteration with U_E
    qc.compose(u_e_operator(params, extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # Measure index register and the first feature qubit
    qc.measure(range(n_qubits_index), range(n_qubits_index))
    qc.measure(n_qubits_index, n_qubits_index)  # measuring the first feature qubit
    
    # Execute circuit
    simulator = Aer.get_backend('qasm_simulator')
    job = execute(qc, simulator, shots=1024)
    counts = job.result().get_counts()
    
    # Target state: index register = (K-1)'s binary representation, and first feature qubit = 1
    target_state = format(K-1, f'0{n_qubits_index}b') + '1'
    p = counts.get(target_state, 0) / 1024.0
    
    # Binary Cross Entropy loss
    epsilon = 1e-10  # 防止 log(0)
    loss = - ( label * np.log(p + epsilon) + (1 - label) * np.log(1 - p + epsilon) )
    return loss

# Training function
def train_gbls(original_dataset, n_epochs=20):
    extended_dataset = build_extended_dataset(original_dataset)
    n_params = 3 * n_qubits_feature * 2  # 3 rotation gates per qubit, 2 layers
    
    # Initialize parameters
    params = np.random.rand(n_params) * 2 * np.pi
    training_loss = []
    
    # Training loop
    for epoch in range(n_epochs):
        np.random.shuffle(extended_dataset)
        epoch_loss = 0
        
        for example in extended_dataset:
            # Define the optimization function for this example
            def opt_func(p):
                return loss_function(p, example)
            
            # Update parameters using COBYLA (with maxiter=5)
            result = minimize(opt_func, params, method='COBYLA', options={'maxiter': 5})
            params = result.x
            epoch_loss += result.fun
        
        avg_loss = epoch_loss / len(extended_dataset)
        training_loss.append(avg_loss)
        print(f"Epoch {epoch+1}/{n_epochs}, Loss: {avg_loss:.4f}")
    
    # Plot training loss
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, n_epochs+1), training_loss)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('GBLS Training Loss (BCE)')
    plt.grid(True)
    plt.show()
    
    return params

# Prediction function as shown in Figure 5
def predict(params, new_input):
    qc = QuantumCircuit(n_qubits_feature, 1)
    # Encode input: apply X for each bit that is '1'
    for i, bit in enumerate(new_input):
        if bit == '1':
            qc.x(i)
    qc.compose(u_l1(params, n_qubits_feature), inplace=True)
    qc.measure(0, 0)
    simulator = Aer.get_backend('qasm_simulator')
    job = execute(qc, simulator, shots=1024)
    counts = job.result().get_counts()
    prob_1 = counts.get('1', 0) / 1024.0
    # 反轉預測結果：若 prob_1 >= 0.5 則返回0，否則返回1
    return 0 if prob_1 >= 0.5 else 1

# Train the model
trained_params = train_gbls(original_dataset)

# Test predictions
for input_data, true_label in original_dataset.items():
    pred = predict(trained_params, input_data)
    print(f"Input: {input_data}, True: {true_label}, Predicted: {pred}")
```

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 定義 3-bit 原始資料集
# -----------------------------
# 這裡根據整數值來設定 label，假設 0~3 為 0，4~7 為 1
original_dataset = { format(i, '03b'): (0 if i < 4 else 1) for i in range(8) }

# Parameters
K = 4  # Extended example 的大小
n_qubits_feature = 3  # 因為3-bit資料，所以需要3個feature qubit
n_qubits_index = 2    # index register 根據 K 保持2個 qubit (因為 2^2 = 4)

# -----------------------------
# Extended dataset 的建構 (見論文描述)
# -----------------------------
def build_extended_dataset(original_data):
    extended_dataset = []
    
    # 分組：分別提取 label 為 0 與 1 的資料
    class0 = [k for k, v in original_data.items() if v == 0]
    class1 = [k for k, v in original_data.items() if v == 1]
    
    # 對於每個原始資料，建立 extended example
    for key, label in original_data.items():
        D_k = []
        # 取與該資料標籤相反的子集合
        opposite_class = class1 if label == 0 else class0
        
        # 加入 K-1 筆對立標籤的例子 (用輪流的方式)
        for i in range(K-1):
            opposite_example = opposite_class[i % len(opposite_class)]
            D_k.append((opposite_example, 1-label))
        
        # 最後加入原始例子
        D_k.append((key, label))
        extended_dataset.append(D_k)
    
    return extended_dataset

# -----------------------------
# U_data: 將 extended example 編碼到量子態
# -----------------------------
def u_data(extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # 對 index register (前 n_qubits_index 個 qubit) 做 Hadamard
    qc.h(range(n_qubits_index))
    
    # 根據 extended example 的每個 pair 進行條件性編碼
    for i, (feature, _) in enumerate(extended_example):
        # 將 i 轉成二進位字串，用來作為控制條件
        bin_i = format(i, f'0{n_qubits_index}b')
        
        # 根據二進位條件設定控制：若 bit 為 '0' 則先翻轉該 qubit
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)
        
        # 對於 feature 字串（此處長度為3），當對應位元為 '1' 時，
        # 對 feature register (從 n_qubits_index 開始) 執行多控制 X 閘
        for j, bit in enumerate(feature):
            if bit == '1':
                qc.mcx(list(range(n_qubits_index)), n_qubits_index + j)
        
        # 還原 index register 上之前做的 X 閘
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)
    
    return qc

# -----------------------------
# U_L1: 參數化變分電路
# -----------------------------
def u_l1(params, n_qubits_feature, layers=2):
    qc = QuantumCircuit(n_qubits_feature)
    param_idx = 0
    
    for _ in range(layers):
        # 每個 qubit 執行 rx, ry, rz 閘
        for i in range(n_qubits_feature):
            qc.rx(params[param_idx], i)
            param_idx += 1
            qc.ry(params[param_idx], i)
            param_idx += 1
            qc.rz(params[param_idx], i)
            param_idx += 1
        
        # 加入循環形式的 CNOT，實現糾纏
        for i in range(n_qubits_feature-1):
            qc.cx(i, i+1)
        qc.cx(n_qubits_feature-1, 0)
    
    return qc

# -----------------------------
# MCZ gate
# -----------------------------
def mcz_gate(n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # 目標 index 為 K-1 (此處 K=4，所以 target_index=3)
    target_index = K-1  
    bin_target = format(target_index, f'0{n_qubits_index}b')
    
    # 對 index register 中 target 為 '0' 的位先應用 X 閘
    for i, bit in enumerate(bin_target):
        if bit == '0':
            qc.x(i)
    
    # 用 index qubit和第一個 feature qubit作為控制，對最後一個 feature qubit施加多控制 Z
    controls = list(range(n_qubits_index)) + [n_qubits_index]
    qc.h(n_qubits_index + n_qubits_feature - 1)  # H 閘前置
    qc.mcx(controls, n_qubits_index + n_qubits_feature - 1)
    qc.h(n_qubits_index + n_qubits_feature - 1)  # H 閘後置
    
    # 還原 index register 的 X 閘
    for i, bit in enumerate(bin_target):
        if bit == '0':
            qc.x(i)
    
    return qc

# -----------------------------
# U_init: Diffusion operator
# -----------------------------
def u_init(n_qubits_index):
    qc = QuantumCircuit(n_qubits_index)
    
    qc.h(range(n_qubits_index))
    qc.x(range(n_qubits_index))
    qc.h(n_qubits_index-1)
    qc.mcx(list(range(n_qubits_index-1)), n_qubits_index-1)
    qc.h(n_qubits_index-1)
    qc.x(range(n_qubits_index))
    qc.h(range(n_qubits_index))
    
    return qc

# -----------------------------
# Complete U operator as defined in Equation (9)
# -----------------------------
def u_operator(params, extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # U_data
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # U_L1
    l1_circuit = u_l1(params, n_qubits_feature)
    qc.compose(l1_circuit, qubits=range(n_qubits_index, n_qubits_index + n_qubits_feature), inplace=True)
    
    # MCZ gate
    qc.compose(mcz_gate(n_qubits_feature, n_qubits_index), inplace=True)
    
    # (U_L1)†
    qc.compose(l1_circuit.inverse(), qubits=range(n_qubits_index, n_qubits_index + n_qubits_feature), inplace=True)
    
    # (U_data)†
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index).inverse(), inplace=True)
    
    # U_init
    qc.compose(u_init(n_qubits_index), qubits=range(n_qubits_index), inplace=True)
    
    return qc

# -----------------------------
# U_E operator for the last cycle
# -----------------------------
def u_e_operator(params, extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # U_data
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # U_L1
    l1_circuit = u_l1(params, n_qubits_feature)
    qc.compose(l1_circuit, qubits=range(n_qubits_index, n_qubits_index + n_qubits_feature), inplace=True)
    
    # MCZ gate
    qc.compose(mcz_gate(n_qubits_feature, n_qubits_index), inplace=True)
    
    # U_init
    qc.compose(u_init(n_qubits_index), qubits=range(n_qubits_index), inplace=True)
    
    return qc

# -----------------------------
# Loss function using Binary Cross Entropy (BCE)
# -----------------------------
def loss_function(params, extended_example):
    # 取得 extended example 最後的對應標籤
    label = extended_example[-1][1]  # label in {0,1}
    
    # 建立 Grover 迭代電路
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index, n_qubits_index + 1)
    
    # 初始 state preparation: 對 index register 做 Hadamard
    qc.h(range(n_qubits_index))
    
    # Grover iterations (根據 O(√K) 次數)
    grover_iterations = int(np.sqrt(K))
    for i in range(grover_iterations - 1):
        qc.compose(u_operator(params, extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # 最後一輪使用 U_E operator
    qc.compose(u_e_operator(params, extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # 測量 index register 與第一個 feature qubit
    qc.measure(range(n_qubits_index), range(n_qubits_index))
    qc.measure(n_qubits_index, n_qubits_index)
    
    # 執行電路
    simulator = Aer.get_backend('qasm_simulator')
    job = execute(qc, simulator, shots=1024)
    counts = job.result().get_counts()
    
    # 目標狀態：index register 為 (K-1) 的二進位形式，加上第一個 feature qubit = '1'
    target_state = format(K-1, f'0{n_qubits_index}b') + '1'
    p = counts.get(target_state, 0) / 1024.0
    
    # BCE loss
    epsilon = 1e-10
    loss = - ( label * np.log(p + epsilon) + (1 - label) * np.log(1 - p + epsilon) )
    return loss

# -----------------------------
# Training function
# -----------------------------
def train_gbls(original_dataset, n_epochs=20):
    extended_dataset = build_extended_dataset(original_dataset)
    n_params = 3 * n_qubits_feature * 2  # 每個 feature qubit 3個旋轉門 * 2 層
    params = np.random.rand(n_params) * 2 * np.pi
    training_loss = []
    
    for epoch in range(n_epochs):
        np.random.shuffle(extended_dataset)
        epoch_loss = 0
        
        for example in extended_dataset:
            def opt_func(p):
                return loss_function(p, example)
            
            result = minimize(opt_func, params, method='COBYLA', options={'maxiter': 5})
            params = result.x
            epoch_loss += result.fun
        
        avg_loss = epoch_loss / len(extended_dataset)
        training_loss.append(avg_loss)
        print(f"Epoch {epoch+1}/{n_epochs}, Loss: {avg_loss:.4f}")
    
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, n_epochs+1), training_loss)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('GBLS Training Loss (BCE)')
    plt.grid(True)
    plt.show()
    
    return params

# -----------------------------
# Prediction function with inverted output
# -----------------------------
def predict(params, new_input):
    qc = QuantumCircuit(n_qubits_feature, 1)
    
    # Encode input: 對 new_input 中為 '1' 的 bit 施加 X 閘
    for i, bit in enumerate(new_input):
        if bit == '1':
            qc.x(i)
    
    # Apply trained U_L1
    qc.compose(u_l1(params, n_qubits_feature), inplace=True)
    
    # Measure the first feature qubit
    qc.measure(0, 0)
    
    simulator = Aer.get_backend('qasm_simulator')
    job = execute(qc, simulator, shots=1024)
    counts = job.result().get_counts()
    prob_1 = counts.get('1', 0) / 1024.0
    
    # 預測時反轉輸出：若 prob_1 >= 0.5，則返回 0；否則返回 1
    return 0 if prob_1 >= 0.5 else 1

# -----------------------------
# 主程式: 訓練模型並進行預測
# -----------------------------
trained_params = train_gbls(original_dataset)

print("\n預測結果：")
for input_data, true_label in original_dataset.items():
    pred = predict(trained_params, input_data)
    print(f"Input: {input_data}, True: {true_label}, Predicted: {pred}")
```

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 定義 4-bit 原始資料集
# -----------------------------
# 產生所有 4-bit 字串，根據整數值設定 label：例如，若 0<= i < 8，label 設為 0，否則設為 1
original_dataset = { format(i, '04b'): (0 if i < 8 else 1) for i in range(16) }

# Parameters
K = 4  # Extended example 的大小
n_qubits_feature = 4  # 4-bit 資料所以 feature register 需要 4 個 qubit
n_qubits_index = 2    # index register 為 2 個 qubit（因為 2^2 = 4，可以表示 K=4 個狀態）

# -----------------------------
# Extended dataset 的建構 (依據論文描述)
# -----------------------------
def build_extended_dataset(original_data):
    extended_dataset = []
    
    # 分類：將 label 為 0 的資料與 label 為 1 的資料分別收集
    class0 = [k for k, v in original_data.items() if v == 0]
    class1 = [k for k, v in original_data.items() if v == 1]
    
    # 對每個原始資料建立 extended example
    for key, label in original_data.items():
        D_k = []
        # 選取與該資料標籤相反的子集合
        opposite_class = class1 if label == 0 else class0
        
        # 加入 K-1 筆對立標籤的例子，若數量不足則用輪流的方式
        for i in range(K-1):
            opposite_example = opposite_class[i % len(opposite_class)]
            D_k.append((opposite_example, 1-label))
        
        # 最後加入原始例子
        D_k.append((key, label))
        extended_dataset.append(D_k)
    
    return extended_dataset

# -----------------------------
# U_data: 將 extended example 的資料編碼至量子態
# -----------------------------
def u_data(extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # 對 index register (前 n_qubits_index 個 qubit) 施加 Hadamard 以產生均勻疊加
    qc.h(range(n_qubits_index))
    
    # 對每個 extended example 中的 pair 進行條件性編碼
    for i, (feature, _) in enumerate(extended_example):
        # 將索引 i 轉換成二進位字串，長度為 n_qubits_index
        bin_i = format(i, f'0{n_qubits_index}b')
        
        # 根據二進位條件對 index register 的 qubit 執行 X 閘（若 bit 為 '0'）
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)
        
        # 根據 feature（長度4）中各位，當位元為 '1' 時，對 feature register (從 n_qubits_index 開始) 執行多控制 X 閘
        for j, bit in enumerate(feature):
            if bit == '1':
                qc.mcx(list(range(n_qubits_index)), n_qubits_index + j)
        
        # 還原剛才對 index register 做的 X 閘
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)
    
    return qc

# -----------------------------
# U_L1: 參數化變分電路，學習特徵空間的超平面
# -----------------------------
def u_l1(params, n_qubits_feature, layers=2):
    qc = QuantumCircuit(n_qubits_feature)
    param_idx = 0
    
    for _ in range(layers):
        # 每個 qubit 執行 rx、ry、rz 閘
        for i in range(n_qubits_feature):
            qc.rx(params[param_idx], i)
            param_idx += 1
            qc.ry(params[param_idx], i)
            param_idx += 1
            qc.rz(params[param_idx], i)
            param_idx += 1
        
        # 循環式的 CNOT 糾纏
        for i in range(n_qubits_feature-1):
            qc.cx(i, i+1)
        qc.cx(n_qubits_feature-1, 0)
    
    return qc

# -----------------------------
# MCZ gate 的實作 (根據論文)
# -----------------------------
def mcz_gate(n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # 目標 index 設為 K-1 (此例中 K=4，所以 target_index = 3)
    target_index = K-1  
    bin_target = format(target_index, f'0{n_qubits_index}b')
    
    # 對 index register 中那些 bit 為 '0' 的位置先施加 X 閘
    for i, bit in enumerate(bin_target):
        if bit == '0':
            qc.x(i)
    
    # 控制條件：所有 index qubit加上第一個 feature qubit作為控制，對最後一個 feature qubit做多控制 Z
    controls = list(range(n_qubits_index)) + [n_qubits_index]
    qc.h(n_qubits_index + n_qubits_feature - 1)  # H 閘作為前置變換
    qc.mcx(controls, n_qubits_index + n_qubits_feature - 1)
    qc.h(n_qubits_index + n_qubits_feature - 1)  # H 閘還原
    
    # 還原 index register 上之前施加的 X 閘
    for i, bit in enumerate(bin_target):
        if bit == '0':
            qc.x(i)
    
    return qc

# -----------------------------
# U_init: Diffusion operator
# -----------------------------
def u_init(n_qubits_index):
    qc = QuantumCircuit(n_qubits_index)
    
    # 實現反射算子 2|ψ⟩⟨ψ| - I，其中 |ψ⟩ 為 index register 上的均勻疊加
    qc.h(range(n_qubits_index))
    qc.x(range(n_qubits_index))
    qc.h(n_qubits_index-1)
    qc.mcx(list(range(n_qubits_index-1)), n_qubits_index-1)
    qc.h(n_qubits_index-1)
    qc.x(range(n_qubits_index))
    qc.h(range(n_qubits_index))
    
    return qc

# -----------------------------
# Complete U operator (Equation (9) in the paper)
# -----------------------------
def u_operator(params, extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # U_data: 正向編碼
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # U_L1: 參數化變分電路，作用在 feature register（位於後面 n_qubits_feature 個 qubit 上）
    l1_circuit = u_l1(params, n_qubits_feature)
    qc.compose(l1_circuit, qubits=range(n_qubits_index, n_qubits_index + n_qubits_feature), inplace=True)
    
    # MCZ gate
    qc.compose(mcz_gate(n_qubits_feature, n_qubits_index), inplace=True)
    
    # (U_L1)† : 逆變分電路
    qc.compose(l1_circuit.inverse(), qubits=range(n_qubits_index, n_qubits_index + n_qubits_feature), inplace=True)
    
    # (U_data)†: 逆向 U_data 編碼
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index).inverse(), inplace=True)
    
    # U_init: diffusion operator (作用在 index register)
    qc.compose(u_init(n_qubits_index), qubits=range(n_qubits_index), inplace=True)
    
    return qc

# -----------------------------
# U_E operator for the last cycle
# -----------------------------
def u_e_operator(params, extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # U_data
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # U_L1
    l1_circuit = u_l1(params, n_qubits_feature)
    qc.compose(l1_circuit, qubits=range(n_qubits_index, n_qubits_index + n_qubits_feature), inplace=True)
    
    # MCZ gate
    qc.compose(mcz_gate(n_qubits_feature, n_qubits_index), inplace=True)
    
    # U_init
    qc.compose(u_init(n_qubits_index), qubits=range(n_qubits_index), inplace=True)
    
    return qc

# -----------------------------
# Loss function: Binary Cross Entropy (BCE) loss
# -----------------------------
def loss_function(params, extended_example):
    # 取得 extended example 最後一筆對應的 label
    label = extended_example[-1][1]  # label 取值 0 或 1
    
    # 建立 Grover 迭代電路
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index, n_qubits_index + 1)
    
    # 初始狀態準備: 對 index register 施加 Hadamard
    qc.h(range(n_qubits_index))
    
    # Grover iterations: 根據 O(√K) 次數
    grover_iterations = int(np.sqrt(K))
    for i in range(grover_iterations - 1):
        qc.compose(u_operator(params, extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # 最後一輪使用 U_E operator
    qc.compose(u_e_operator(params, extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    # 測量 index register 與第一個 feature qubit
    qc.measure(range(n_qubits_index), range(n_qubits_index))
    qc.measure(n_qubits_index, n_qubits_index)
    
    # 執行電路
    simulator = Aer.get_backend('qasm_simulator')
    job = execute(qc, simulator, shots=1024)
    counts = job.result().get_counts()
    
    # 目標狀態定義：index register = (K-1) 的二進位形式，加上第一個 feature qubit = '1'
    target_state = format(K-1, f'0{n_qubits_index}b') + '1'
    p = counts.get(target_state, 0) / 1024.0
    
    # 計算 BCE loss
    epsilon = 1e-10
    loss = - ( label * np.log(p + epsilon) + (1 - label) * np.log(1 - p + epsilon) )
    return loss

# -----------------------------
# Training function
# -----------------------------
def train_gbls(original_dataset, n_epochs=20):
    extended_dataset = build_extended_dataset(original_dataset)
    # 參數數量：每個 feature qubit 3 個旋轉門 * 2 層
    n_params = 3 * n_qubits_feature * 2  
    params = np.random.rand(n_params) * 2 * np.pi
    training_loss = []
    
    for epoch in range(n_epochs):
        np.random.shuffle(extended_dataset)
        epoch_loss = 0
        
        for example in extended_dataset:
            def opt_func(p):
                return loss_function(p, example)
            result = minimize(opt_func, params, method='COBYLA', options={'maxiter': 5})
            params = result.x
            epoch_loss += result.fun
        
        avg_loss = epoch_loss / len(extended_dataset)
        training_loss.append(avg_loss)
        print(f"Epoch {epoch+1}/{n_epochs}, Loss: {avg_loss:.4f}")
    
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, n_epochs+1), training_loss)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('GBLS Training Loss (BCE) for 4-bit Dataset')
    plt.grid(True)
    plt.show()
    
    return params

# -----------------------------
# Prediction function with inverted output
# -----------------------------
def predict(params, new_input):
    qc = QuantumCircuit(n_qubits_feature, 1)
    
    # 輸入編碼：對 new_input 中為 '1' 的位元施加 X 閘
    for i, bit in enumerate(new_input):
        if bit == '1':
            qc.x(i)
    
    # 使用訓練後的 U_L1 變分電路
    qc.compose(u_l1(params, n_qubits_feature), inplace=True)
    
    # 測量第一個 feature qubit
    qc.measure(0, 0)
    
    simulator = Aer.get_backend('qasm_simulator')
    job = execute(qc, simulator, shots=1024)
    counts = job.result().get_counts()
    prob_1 = counts.get('1', 0) / 1024.0
    
    # 預測時反轉輸出：若 prob_1 >= 0.5 則返回 0；否則返回 1
    return 0 if prob_1 >= 0.5 else 1

# -----------------------------
# 主程式：訓練模型並進行預測
# -----------------------------
trained_params = train_gbls(original_dataset, n_epochs=20)

print("\n預測結果：")
for input_data, true_label in original_dataset.items():
    pred = predict(trained_params, input_data)
    print(f"Input: {input_data}, True: {true_label}, Predicted: {pred}")
```

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt

#############################
# 資料分割工具函數
#############################
def split_dataset(original, train_ratio=0.5, vali_ratio=0.25, test_ratio=0.25):
    """將原始資料集（字典）隨機分割成訓練、驗證和測試三部分"""
    keys = list(original.keys())
    np.random.shuffle(keys)
    n = len(keys)
    n_train = int(train_ratio * n)
    n_vali = int(vali_ratio * n)
    train_keys = keys[:n_train]
    vali_keys = keys[n_train:n_train+n_vali]
    test_keys = keys[n_train+n_vali:]
    train_data = {k: original[k] for k in train_keys}
    vali_data = {k: original[k] for k in vali_keys}
    test_data  = {k: original[k] for k in test_keys}
    return train_data, vali_data, test_data

#############################
# 定義 6-bit 原始資料集
#############################
# 產生所有 6-bit 字串（"000000" ～ "111111"），共 64 筆，
# 根據整數值設定 label：當 i < 32 時 label 為0；否則為1
original_dataset = { format(i, '06b'): (0 if i < 32 else 1) for i in range(64) }

# 分割資料集（例如：50% 訓練, 25% 驗證, 25% 測試）
train_dataset, vali_dataset, test_dataset = split_dataset(original_dataset, train_ratio=0.5, vali_ratio=0.25, test_ratio=0.25)
print("Train dataset sample (6-bit):", list(train_dataset.items())[:5])
print("Validation dataset sample (6-bit):", list(vali_dataset.items())[:5])
print("Test dataset sample (6-bit):", list(test_dataset.items())[:5])

#############################
# 參數設定
#############################
K = 4                        # Extended example 的大小
n_qubits_feature = 6         # 每筆資料為 6-bit，因此 feature register 使用 6 個 qubit
n_qubits_index = 2           # index register 為 2 個 qubit (2^2=4, 可表示 K=4 個索引)

#############################
# Extended dataset 建構 (依據論文描述)
#############################
def build_extended_dataset(dataset):
    """根據傳入的 dataset（字典）構造 extended dataset，每個 extended example D_k
       包含前 K-1 筆對立例子與最後一筆原始例子。"""
    extended_dataset = []
    
    # 分組：將標籤為 0 與標籤為 1 的資料分開收集
    class0 = [k for k, v in dataset.items() if v == 0]
    class1 = [k for k, v in dataset.items() if v == 1]
    
    for key, label in dataset.items():
        D_k = []
        opposite_class = class1 if label == 0 else class0
        
        # 加入前 K-1 筆對立例子，若不足則輪流取樣
        for i in range(K-1):
            opposite_example = opposite_class[i % len(opposite_class)]
            D_k.append((opposite_example, 1 - label))
        
        # 加入原始例子
        D_k.append((key, label))
        extended_dataset.append(D_k)
    
    return extended_dataset

#############################
# 各子電路模組 (基本與之前相同)
#############################
def u_data(extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    
    # 對 index register (前 n_qubits_index 個 qubit) 施加 Hadamard
    qc.h(range(n_qubits_index))
    
    # 對每個 extended example 中的 pair 做條件性編碼
    for i, (feature, _) in enumerate(extended_example):
        # 將 i 轉成二進位字串，長度為 n_qubits_index
        bin_i = format(i, f'0{n_qubits_index}b')
        # 根據條件對 index register 做 X 閘
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)
        # 對 feature (6-bit) 中為 '1' 的位置，對 feature register (從 n_qubits_index 開始) 施加多控制 X
        for j, bit in enumerate(feature):
            if bit == '1':
                qc.mcx(list(range(n_qubits_index)), n_qubits_index + j)
        # 還原 index register 的 X 閘
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)
    return qc

def u_l1(params, n_qubits_feature, layers=2):
    qc = QuantumCircuit(n_qubits_feature)
    param_idx = 0
    for _ in range(layers):
        for i in range(n_qubits_feature):
            qc.rx(params[param_idx], i)
            param_idx += 1
            qc.ry(params[param_idx], i)
            param_idx += 1
            qc.rz(params[param_idx], i)
            param_idx += 1
        for i in range(n_qubits_feature - 1):
            qc.cx(i, i+1)
        qc.cx(n_qubits_feature-1, 0)
    return qc

def mcz_gate(n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    target_index = K - 1  
    bin_target = format(target_index, f'0{n_qubits_index}b')
    for i, bit in enumerate(bin_target):
        if bit == '0':
            qc.x(i)
    controls = list(range(n_qubits_index)) + [n_qubits_index]
    qc.h(n_qubits_index + n_qubits_feature - 1)
    qc.mcx(controls, n_qubits_index + n_qubits_feature - 1)
    qc.h(n_qubits_index + n_qubits_feature - 1)
    for i, bit in enumerate(bin_target):
        if bit == '0':
            qc.x(i)
    return qc

def u_init(n_qubits_index):
    qc = QuantumCircuit(n_qubits_index)
    qc.h(range(n_qubits_index))
    qc.x(range(n_qubits_index))
    qc.h(n_qubits_index-1)
    qc.mcx(list(range(n_qubits_index-1)), n_qubits_index-1)
    qc.h(n_qubits_index-1)
    qc.x(range(n_qubits_index))
    qc.h(range(n_qubits_index))
    return qc

def u_operator(params, extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    l1_circuit = u_l1(params, n_qubits_feature)
    qc.compose(l1_circuit, qubits=range(n_qubits_index, n_qubits_index+n_qubits_feature), inplace=True)
    qc.compose(mcz_gate(n_qubits_feature, n_qubits_index), inplace=True)
    qc.compose(l1_circuit.inverse(), qubits=range(n_qubits_index, n_qubits_index+n_qubits_feature), inplace=True)
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index).inverse(), inplace=True)
    qc.compose(u_init(n_qubits_index), qubits=range(n_qubits_index), inplace=True)
    return qc

def u_e_operator(params, extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    qc.compose(u_data(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    l1_circuit = u_l1(params, n_qubits_feature)
    qc.compose(l1_circuit, qubits=range(n_qubits_index, n_qubits_index+n_qubits_feature), inplace=True)
    qc.compose(mcz_gate(n_qubits_feature, n_qubits_index), inplace=True)
    qc.compose(u_init(n_qubits_index), qubits=range(n_qubits_index), inplace=True)
    return qc

#############################
# Loss function: 使用 BCE loss
#############################
def loss_function(params, extended_example):
    # 取得 extended example 最後一筆例子的 label
    label = extended_example[-1][1]
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index, n_qubits_index + 1)
    qc.h(range(n_qubits_index))
    grover_iterations = int(np.sqrt(K))
    for i in range(grover_iterations - 1):
        qc.compose(u_operator(params, extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    qc.compose(u_e_operator(params, extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    qc.measure(range(n_qubits_index), range(n_qubits_index))
    qc.measure(n_qubits_index, n_qubits_index)
    simulator = Aer.get_backend('qasm_simulator')
    job = execute(qc, simulator, shots=1024)
    counts = job.result().get_counts()
    target_state = format(K-1, f'0{n_qubits_index}b') + '1'
    p = counts.get(target_state, 0) / 1024.0
    epsilon = 1e-10
    loss = - ( label * np.log(p + epsilon) + (1 - label) * np.log(1 - p + epsilon) )
    return loss

#############################
# Training function
#############################
def train_gbls(dataset, n_epochs=20):
    extended_dataset = build_extended_dataset(dataset)
    n_params = 3 * n_qubits_feature * 2  # 每個 feature qubit 3 個旋轉門 * 2 層
    params = np.random.rand(n_params) * 2 * np.pi
    training_loss = []
    for epoch in range(n_epochs):
        np.random.shuffle(extended_dataset)
        epoch_loss = 0
        for example in extended_dataset:
            def opt_func(p):
                return loss_function(p, example)
            result = minimize(opt_func, params, method='COBYLA', options={'maxiter': 5})
            params = result.x
            epoch_loss += result.fun
        avg_loss = epoch_loss / len(extended_dataset)
        training_loss.append(avg_loss)
        print(f"Epoch {epoch+1}/{n_epochs}, Loss: {avg_loss:.4f}")
    plt.figure(figsize=(10,6))
    plt.plot(range(1, n_epochs+1), training_loss)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('GBLS Training Loss (BCE) - 6-bit Train Set')
    plt.grid(True)
    plt.show()
    return params

#############################
# Prediction function (反轉輸出)
#############################
def predict(params, new_input):
    qc = QuantumCircuit(n_qubits_feature, 1)
    for i, bit in enumerate(new_input):
        if bit == '1':
            qc.x(i)
    qc.compose(u_l1(params, n_qubits_feature), inplace=True)
    qc.measure(0, 0)
    simulator = Aer.get_backend('qasm_simulator')
    job = execute(qc, simulator, shots=1024)
    counts = job.result().get_counts()
    prob_1 = counts.get('1', 0) / 1024.0
    # 反轉預測：若 prob_1 >= 0.5 則返回 0，否則返回 1
    return 0 if prob_1 >= 0.5 else 1

#############################
# 主程式：分割資料、訓練並評估
#############################
train_data, vali_data, test_data = split_dataset(original_dataset, train_ratio=0.5, vali_ratio=0.25, test_ratio=0.25)
print("Train dataset sample (6-bit):", list(train_data.items())[:5])
print("Validation dataset sample (6-bit):", list(vali_data.items())[:5])
print("Test dataset sample (6-bit):", list(test_data.items())[:5])

trained_params = train_gbls(train_data, n_epochs=20)

def evaluate(dataset, params):
    correct = 0
    for key, true_label in dataset.items():
        pred = predict(params, key)
        if pred == true_label:
            correct += 1
    return correct / len(dataset)

train_acc = evaluate(train_data, trained_params)
vali_acc  = evaluate(vali_data, trained_params)
test_acc  = evaluate(test_data, trained_params)

print("\n評估結果：")
print("Train Accuracy:", train_acc)
print("Validation Accuracy:", vali_acc)
print("Test Accuracy:", test_acc)

print("\n詳細預測：")
for ds_name, ds in zip(["Train", "Validation", "Test"], [train_data, vali_data, test_data]):
    print(f"{ds_name} predictions:")
    for key, true_label in ds.items():
        pred = predict(trained_params, key)
        print(f"Input: {key}, True: {true_label}, Predicted: {pred}")
```

```python
import numpy as np
import csv
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram


K = 4
##########################################
# (一) Synthetic Dataset 產生模組
##########################################
def random_su4(seed=None):
    """
    產生一個 4x4 隨機 SU(4) 單位矩陣
    """
    if seed is not None:
        np.random.seed(seed)
    real_part = np.random.randn(4,4)
    imag_part = np.random.randn(4,4)
    z = real_part + 1j*imag_part
    q, r = np.linalg.qr(z)
    det_q = np.linalg.det(q)
    q = q * (det_q/np.abs(det_q))**(-1/4)
    return q

def feature_map_g(omega1, omega2):
    """
    非線性特徵嵌入 g(ω1,ω2) ∈ ℂ^4
    實作示例：
      g(ω1, ω2) = [cos(ω1 - ω2), sin(ω1 - ω2), cos((ω1-ω2)^2), sin((ω1-ω2)^2)]
    """
    diff = omega1 - omega2
    # 加入一個倍率因子，例如使 (ω1-ω2)^2 部分更敏感
    alpha = 2.0  
    return np.array([
        np.cos(diff),
        np.sin(diff),
        np.cos(alpha * diff**2),
        np.sin(alpha * diff**2)
    ], dtype=complex)

def measure_overlap(vec, V, Pi):
    """
    計算 <vec|V^\dagger Pi V|vec>，作為資料 label 的依據
    """
    state = V @ vec
    mid = Pi @ state
    val = np.vdot(state, mid)
    return np.real(val)

def generate_synthetic_dataset(N=200, delta=0.2, seed=42, discard_inbetween=False):
    """
    根據論文生成合成資料集
    每筆資料 x = (ω1, ω2), ω1,ω2 均服從 Uniform(0,2π)
    並依據 <g(x)|V^† Π V |g(x)> 與 delta 判斷 label
    回傳：X (shape=(N,2)) 與 Y (shape=(N,))
    """
    if seed is not None:
        np.random.seed(seed)
    V = random_su4()
    Pi = np.diag([1, 0, 1, 0])  # 定義測量算符
    X_data = []
    Y_data = []
    count = 0
    while len(X_data) < N:
        w1 = 2*np.pi*np.random.rand()
        w2 = 2*np.pi*np.random.rand()
        vec = feature_map_g(w1, w2)
        norm = np.linalg.norm(vec)
        if norm > 1e-12:
            vec /= norm
        val = measure_overlap(vec, V, Pi)
        if val >= 0.5 + delta:
            label = 1
        elif val <= 0.5 - delta:
            label = 0
        else:
            if discard_inbetween:
                continue
            else:
                label = 1 if val >= 0.5 else 0
        X_data.append([w1, w2])
        Y_data.append(label)
        count += 1
    X_arr = np.array(X_data[:N])
    Y_arr = np.array(Y_data[:N])
    print(f"完成生成 N={N} 筆資料 (嘗試 {count} 次隨機)。")
    return X_arr, Y_arr

##########################################
# (二) 將 synthetic dataset 轉成 GBLS 使用的格式
##########################################
def build_extended_dataset_synthetic(X, Y, K=4):
    """
    將 synthetic dataset (X, Y) 轉換成 GBLS 需要的 extended dataset 格式。
    每個樣本為 tuple: (w1, w2, label)
    Extended example D_k 為長度 K 的 list，前 K-1 筆來自與原樣本 label 相反的子集合，最後一筆為原樣本
    回傳：extended_dataset 為 list，每個元素對應一個 extended sample (list of tuples)
    """
    samples = [(X[i,0], X[i,1], Y[i]) for i in range(len(Y))]
    extended_dataset = []
    # 將 samples 分為兩組
    class0 = [s for s in samples if s[2]==0]
    class1 = [s for s in samples if s[2]==1]
    for sample in samples:
        label = sample[2]
        ext = []
        opposite = class1 if label==0 else class0
        for i in range(K-1):
            ext.append(opposite[i % len(opposite)])
        ext.append(sample)
        extended_dataset.append(ext)
    return extended_dataset

##########################################
# (三) GBLS 量子電路模組（基於 synthetic dataset 的 encoding）
##########################################
# 在 synthetic dataset 下，我們希望用 feature_map_g 將 (ω1, ω2) 映射成 4-dim state，
# 因此我們設定 feature register 為 2 qubits（2^2 = 4）。
n_qubits_feature = 2  # 注意：這裡固定為 2，因為 feature_map_g 輸出 4 維向量
n_qubits_index = 2    # K=4，所以 index register 仍用 2 個 qubit

def u_data_synthetic(extended_example, n_qubits_feature, n_qubits_index):
    """
    將 synthetic extended sample 進行 state preparation。
    每筆 extended example 中，每個元素是 (w1, w2, label)。
    這裡利用 initialize 將 feature_map_g(ω1,ω2) 對應的 state 準備到 feature register（佔用 qubits [n_qubits_index, ...]）。
    為了模仿原 GBLS 結構，對於每個 extended pair以其在 extended sample 中的索引 i 設定控制（透過 X 閘）。
    """
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    qc.h(range(n_qubits_index))
    # 定義一個小函數，確保轉為 Python 原生數值
    def convert_vector(vec):
        return [float(x.real) if abs(x.imag) < 1e-12 else complex(x) for x in vec.tolist()]
    for i, (w1, w2, _) in enumerate(extended_example):
        bin_i = format(i, f'0{n_qubits_index}b')
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)
        vec = feature_map_g(w1, w2)
        norm = np.linalg.norm(vec)
        if norm > 1e-12:
            vec = vec / norm
        vec_list = convert_vector(vec)
        feature_qubits = list(range(n_qubits_index, n_qubits_index+n_qubits_feature))
        # 將 feature state 初始化到 feature register
        qc.initialize(vec_list, feature_qubits)
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)
    return qc

def u_data_synthetic_inverse(extended_example, n_qubits_feature, n_qubits_index):
    """
    Manual inverse implementation of u_data_synthetic
    """
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    # Apply H gates in reverse
    qc.h(range(n_qubits_index))
    
    # Process the extended_example in reverse order
    for i in range(len(extended_example)-1, -1, -1):
        w1, w2, _ = extended_example[i]
        bin_i = format(i, f'0{n_qubits_index}b')
        
        # Apply X gates if needed
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)
        
        # For the inverse of initialization, we need to initialize with the complex conjugate
        vec = feature_map_g(w1, w2)
        norm = np.linalg.norm(vec)
        if norm > 1e-12:
            vec = vec / norm
            
        # Take complex conjugate for the inverse
        vec_conj = np.conjugate(vec)
        
        # Convert to list with proper handling of complex numbers
        vec_list = [float(x.real) if abs(x.imag) < 1e-12 else complex(x) for x in vec_conj.tolist()]
        
        # Initialize with conjugate state
        feature_qubits = list(range(n_qubits_index, n_qubits_index+n_qubits_feature))
        qc.initialize(vec_list, feature_qubits)
        
        # Reverse X gates if needed
        for j, bit in enumerate(bin_i):
            if bit == '0':
                qc.x(j)
    
    return qc

def u_l1(params, n_qubits_feature, layers=2):
    qc = QuantumCircuit(n_qubits_feature)
    param_idx = 0
    for _ in range(layers):
        for i in range(n_qubits_feature):
            qc.rx(params[param_idx], i)
            param_idx += 1
            qc.ry(params[param_idx], i)
            param_idx += 1
            qc.rz(params[param_idx], i)
            param_idx += 1
        for i in range(n_qubits_feature-1):
            qc.cx(i, i+1)
        qc.cx(n_qubits_feature-1, 0)
    return qc

def mcz_gate(n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    target_index = K - 1
    bin_target = format(target_index, f'0{n_qubits_index}b')
    for i, bit in enumerate(bin_target):
        if bit == '0':
            qc.x(i)
    controls = list(range(n_qubits_index)) + [n_qubits_index]
    qc.h(n_qubits_index + n_qubits_feature - 1)
    qc.mcx(controls, n_qubits_index + n_qubits_feature - 1)
    qc.h(n_qubits_index + n_qubits_feature - 1)
    for i, bit in enumerate(bin_target):
        if bit == '0':
            qc.x(i)
    return qc

def u_init(n_qubits_index):
    qc = QuantumCircuit(n_qubits_index)
    qc.h(range(n_qubits_index))
    qc.x(range(n_qubits_index))
    qc.h(n_qubits_index-1)
    qc.mcx(list(range(n_qubits_index-1)), n_qubits_index-1)
    qc.h(n_qubits_index-1)
    qc.x(range(n_qubits_index))
    qc.h(range(n_qubits_index))
    return qc

def u_operator_synthetic(params, extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    qc.compose(u_data_synthetic(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    l1_circuit = u_l1(params, n_qubits_feature)
    qc.compose(l1_circuit, qubits=range(n_qubits_index, n_qubits_index+n_qubits_feature), inplace=True)
    qc.compose(mcz_gate(n_qubits_feature, n_qubits_index), inplace=True)
    qc.compose(l1_circuit.inverse(), qubits=range(n_qubits_index, n_qubits_index+n_qubits_feature), inplace=True)
    
    # Use our manual inverse implementation instead of .inverse()
    qc.compose(u_data_synthetic_inverse(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    
    qc.compose(u_init(n_qubits_index), qubits=range(n_qubits_index), inplace=True)
    return qc

def u_e_operator_synthetic(params, extended_example, n_qubits_feature, n_qubits_index):
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index)
    qc.compose(u_data_synthetic(extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    l1_circuit = u_l1(params, n_qubits_feature)
    qc.compose(l1_circuit, qubits=range(n_qubits_index, n_qubits_index+n_qubits_feature), inplace=True)
    qc.compose(mcz_gate(n_qubits_feature, n_qubits_index), inplace=True)
    qc.compose(u_init(n_qubits_index), qubits=range(n_qubits_index), inplace=True)
    return qc

##########################################
# (四) Loss, Training 與 Prediction 模組 (基於 synthetic dataset)
##########################################
def loss_function_synthetic(params, extended_example):
    label = extended_example[-1][2]  # 取得 extended example 中最後一筆的 label
    qc = QuantumCircuit(n_qubits_feature + n_qubits_index, n_qubits_index + 1)
    qc.h(range(n_qubits_index))
    grover_iterations = int(np.sqrt(K))
    for i in range(grover_iterations - 1):
        qc.compose(u_operator_synthetic(params, extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    qc.compose(u_e_operator_synthetic(params, extended_example, n_qubits_feature, n_qubits_index), inplace=True)
    qc.measure(range(n_qubits_index), range(n_qubits_index))
    qc.measure(n_qubits_index, n_qubits_index)
    simulator = Aer.get_backend('qasm_simulator')
    job = execute(qc, simulator, shots=1024)
    counts = job.result().get_counts()
    target_state = format(K-1, f'0{n_qubits_index}b') + '1'
    p = counts.get(target_state, 0) / 1024.0
    epsilon = 1e-10
    loss = - ( label * np.log(p + epsilon) + (1 - label) * np.log(1 - p + epsilon) )
    return loss

def train_gbls_synthetic(X, Y, n_epochs=20):
    extended_dataset = build_extended_dataset_synthetic(X, Y, K=K)
    n_params = 3 * n_qubits_feature * 2  # 3 rotation gates per feature qubit, 2 layers
    params = np.random.rand(n_params) * 2 * np.pi
    training_loss = []
    for epoch in range(n_epochs):
        np.random.shuffle(extended_dataset)
        epoch_loss = 0
        for ext in extended_dataset:
            def opt_func(p):
                return loss_function_synthetic(p, ext)
            result = minimize(opt_func, params, method='COBYLA', options={'maxiter': 5})
            params = result.x
            epoch_loss += result.fun
        avg_loss = epoch_loss / len(extended_dataset)
        training_loss.append(avg_loss)
        print(f"Epoch {epoch+1}/{n_epochs}, Loss: {avg_loss:.4f}")
    plt.figure(figsize=(10,6))
    plt.plot(range(1, n_epochs+1), training_loss)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('GBLS Training Loss (BCE) - Synthetic  Dataset')
    plt.grid(True)
    plt.show()
    return params

def predict_synthetic(params, sample):
    # sample 為 (w1, w2, label) 其中 label 可忽略；我們只用 (w1, w2) 進行 encoding
    # 這裡只執行 U_L1，並測量第一個 feature qubit，然後反轉輸出
    qc = QuantumCircuit(n_qubits_feature, 1)
    # 直接使用 initialize 運行 feature_map_g 將 sample 轉 state
    vec = feature_map_g(sample[0], sample[1])
    norm = np.linalg.norm(vec)
    if norm > 1e-12:
        vec = vec / norm
    def convert_vector(vec):
        return [float(x.real) if abs(x.imag) < 1e-12 else complex(x) for x in vec.tolist()]
    vec_list = convert_vector(vec)
    qc.initialize(vec_list, list(range(n_qubits_feature)))
    qc.compose(u_l1(params, n_qubits_feature), inplace=True)
    qc.measure(0, 0)
    simulator = Aer.get_backend('qasm_simulator')
    job = execute(qc, simulator, shots=1024)
    counts = job.result().get_counts()
    prob_1 = counts.get('1', 0) / 1024.0
    return 0 if prob_1 < 0.5 else 1

def evaluate_synthetic(dataset, params):
    # dataset 為 list of samples (w1, w2, label)
    correct = 0
    for sample in dataset:
        pred = predict_synthetic(params, sample)
        if pred == sample[2]:
            correct += 1
    return correct / len(dataset)

##########################################
# (五) 主程式：生成資料、訓練並評估
##########################################
if __name__ == "__main__":
    # 生成 synthetic dataset (例如 N=200)
    X, Y = generate_synthetic_dataset(N=200, delta=0.2, seed=123, discard_inbetween=True)
    print("前 10 筆 synthetic 資料:")
    for i in range(10):
        print(f"{i}: x={X[i]}, y={Y[i]}")
    
    # 建立 synthetic dataset 為 list of samples：每個 sample = (w1, w2, label)
    synthetic_samples = [(X[i,0], X[i,1], Y[i]) for i in range(len(Y))]
    
    # 分割 synthetic dataset（例如 50% 訓練，25% 驗證，25% 測試）
    def split_samples(samples, train_ratio=0.5, vali_ratio=0.25, test_ratio=0.25):
        np.random.shuffle(samples)
        n = len(samples)
        n_train = int(train_ratio * n)
        n_vali = int(vali_ratio * n)
        train = samples[:n_train]
        vali = samples[n_train:n_train+n_vali]
        test = samples[n_train+n_vali:]
        return train, vali, test
    train_samples, vali_samples, test_samples = split_samples(synthetic_samples)
    print("\n訓練集 sample:", train_samples[:5])
    print("驗證集 sample:", vali_samples[:5])
    print("測試集 sample:", test_samples[:5])
    
    # 訓練 GBLS 模型（基於 synthetic dataset）
    trained_params = train_gbls_synthetic(X, Y, n_epochs=20)
    
    # 評估訓練、驗證、測試準確率
    train_acc = evaluate_synthetic(train_samples, trained_params)
    vali_acc = evaluate_synthetic(vali_samples, trained_params)
    test_acc = evaluate_synthetic(test_samples, trained_params)
    print("\n評估結果：")
    print("Train Accuracy:", train_acc)
    print("Validation Accuracy:", vali_acc)
    print("Test Accuracy:", test_acc)
    
    print("\n詳細預測：")
    for ds_name, ds in zip(["Train", "Validation", "Test"], [train_samples, vali_samples, test_samples]):
        print(f"{ds_name} predictions:")
        for sample in ds:
            pred = predict_synthetic(trained_params, sample)
            print(f"Input: ({sample[0]:.3f}, {sample[1]:.3f}), True: {sample[2]}, Predicted: {pred}")
```

```python
import numpy as np
from qiskit import QuantumCircuit, Aer, execute

# 假設以下函式與常數已在同檔中定義，或可從你的 GBLS 模組匯入：
# - generate_synthetic_dataset, build_extended_dataset_synthetic
# - u_operator_synthetic, u_e_operator_synthetic
# - n_qubits_feature, n_qubits_index, K

# --------------------
# (A) QOTP Pauli One-Time-Pad
# --------------------
def qotp_encrypt(qc: QuantumCircuit):
    """
    在 qc 上對每個 qubit 施加隨機 X^a Z^b 並回傳 key list
    key list 格式: [(a0,b0), (a1,b1), ...]
    """
    keys = []
    for idx, q in enumerate(qc.qubits):
        a = np.random.randint(2)
        b = np.random.randint(2)
        if b:
            qc.z(q)
        if a:
            qc.x(q)
        keys.append((a, b))
    return keys


def qotp_update_keys(keys, gate, qubits):
    """
    根據雲端在加密態上執行的 gate 更新本地 key
      - gate: 'h', 's', 'x', 'z', 'cx'
      - qubits: [i] or [ctrl, tgt]
    """
    if gate == 'h':
        i = qubits[0]
        a, b = keys[i]
        keys[i] = (b, a)
    elif gate == 's':
        i = qubits[0]
        a, b = keys[i]
        keys[i] = (a, a ^ b)
    elif gate == 'cx':
        c, t = qubits
        a_c, b_c = keys[c]
        a_t, b_t = keys[t]
        keys[t] = (a_t ^ a_c, b_t ^ b_c)
    # X, Z gates commute up to global phase: key 不變
    return keys


def qotp_decrypt(qc: QuantumCircuit, keys):
    """
    依序對每個 qubit 施加解密 X^a Z^b
    """
    for idx, (a, b) in enumerate(keys):
        if a:
            qc.x(qc.qubits[idx])
        if b:
            qc.z(qc.qubits[idx])
    return qc

# --------------------
# (B) 雲端 GBLS 分類：將 u_operator_synthetic 與 u_e_operator_synthetic 串接進去
# --------------------
def server_gbls_classification(qc: QuantumCircuit, params, extended_example):
    """
    在加密後的 qc 上執行 GBLS Grover 迭代 + oracle 回傳完整 gate log
    """
    gate_log = []
    # Grover 迭代次數 ~ sqrt(K)
    n_iter = int(np.sqrt(K))
    # 前 n_iter-1 次跑完整 operator
    for _ in range(n_iter - 1):
        op = u_operator_synthetic(params, extended_example, n_qubits_feature, n_qubits_index)
        qc.compose(op, inplace=True)
        for inst, qargs, _ in op.data:
            gate_log.append((inst.name, [q.index for q in qargs]))
    # 最後一次用 u_e_operator_synthetic
    op_e = u_e_operator_synthetic(params, extended_example, n_qubits_feature, n_qubits_index)
    qc.compose(op_e, inplace=True)
    for inst, qargs, _ in op_e.data:
        gate_log.append((inst.name, [q.index for q in qargs]))
    return gate_log

# --------------------
# (C) 主流程示例：資料生成 → 加密 → 上傳雲端 → 解密 → 量測
# --------------------
if __name__ == '__main__':
    # 1. 產生合成資料與 extended examples
    X, Y = generate_synthetic_dataset(N=100, delta=0.2, seed=42, discard_inbetween=True)
    extended_dataset = build_extended_dataset_synthetic(X, Y, K)
    # 取第一筆 extended example
    example = extended_dataset[0]

    # 2. 分類模型參數 (可換成訓練好後的參數)
    params = np.random.rand(3 * n_qubits_feature * 2) * 2 * np.pi

    # 3. 本地端：建立空 circuit，並做 QOTP 加密
    qc = QuantumCircuit(n_qubits_index + n_qubits_feature)
    keys = qotp_encrypt(qc)

    # 4. 上傳雲端：在加密態上執行 GBLS 分類 (Grover 迭代)
    gate_log = server_gbls_classification(qc, params, example)

    # 5. 本地端：根據 gate_log 更新 key，然後做解密
    for gate, qs in gate_log:
        keys = qotp_update_keys(keys, gate, qs)
    qc = qotp_decrypt(qc, keys)

    # 6. 本地端：量測並讀取結果
    qc.measure_all()
    sim = Aer.get_backend('qasm_simulator')
    counts = execute(qc, sim, shots=1024).result().get_counts()
    print('Classification counts:', counts)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def random_su4(seed=None):
    if seed is not None:
        np.random.seed(seed)
    real_part = np.random.randn(4,4)
    imag_part = np.random.randn(4,4)
    z = real_part + 1j*imag_part
    q, r = np.linalg.qr(z)
    det_q = np.linalg.det(q)
    q = q * (det_q/np.abs(det_q))**(-1/4)
    return q

def feature_map_g(omega1, omega2):
    """
    非線性特徵嵌入 g(ω1,ω2) ∈ ℂ^4
    實作示例：
      g(ω1, ω2) = [cos(ω1 - ω2), sin(ω1 - ω2), cos((ω1-ω2)^2), sin((ω1-ω2)^2)]
    """
    diff = omega1 - omega2
    # 加入一個倍率因子，例如使 (ω1-ω2)^2 部分更敏感
    alpha = 2.0  
    return np.array([
        np.cos(diff),
        np.sin(diff),
        np.cos(alpha * diff**2),
        np.sin(alpha * diff**2)
    ], dtype=complex)

def measure_overlap(vec, V, Pi):
    state = V @ vec
    mid = Pi @ state
    val = np.vdot(state, mid)  # 自動對複數取共軛
    return np.real(val)

# 設置參數
N = 1000             # 取1000個樣本檢查分佈
delta = 0.2          # 與論文中設定相同
V = random_su4(seed=123)
Pi = np.diag([1, 0, 1, 0])  # Π = I ⊗ |0><0|, 對應4維向量

vals = []
for _ in range(N):
    w1 = 2 * np.pi * np.random.rand()
    w2 = 2 * np.pi * np.random.rand()
    vec = feature_map_g(w1, w2)
    norm = np.linalg.norm(vec)
    if norm > 1e-12:
        vec /= norm
    v_val = measure_overlap(vec, V, Pi)
    vals.append(v_val)

vals = np.array(vals)

# 畫出直方圖
plt.figure(figsize=(8, 5))
plt.hist(vals, bins=50, edgecolor='black', alpha=0.75)
plt.xlabel(r"$\langle g(x)|V^\dagger \Pi V|g(x) \rangle$")
plt.ylabel("Samples")
plt.title(r"Distribution of $\langle g(x)|V^\dagger \Pi V|g(x) \rangle$")
plt.axvline(0.5 - delta, color='red', linestyle='dashed', label=f"0.5 - δ = {0.5-delta}")
plt.axvline(0.5 + delta, color='green', linestyle='dashed', label=f"0.5 + δ = {0.5+delta}")
plt.legend()
plt.show()
```
